```python
import numpy as np 
from matplotlib import pyplot as plt 
import cv2 as cv
```

```python
s = 500
n = 1500
xs = np.floor(np.random.random((n))*s).astype('int')
ys = np.floor(np.random.random((n))*s).astype('int')
rs = np.floor(np.random.random((n))*0.2*s).astype('int')

pts = np.array([xs,ys,rs])
```

```python
data_set = np.ones((n,s,s),dtype=np.float16)
                
```

```python
i = 0
for c in data_set: 
    cv.circle(img=c,center=(xs[i],ys[i]),radius=rs[i],color=(128),thickness=2)
    i+=1
```

```python
data_set = data_set.reshape(n,-1).transpose()
```

```python
img = cv.imread('pic1.jpg')
```

```python
img = cv.resize(img,(s,s))
```

```python
img = img[:,:,0]
```

```python
from sklearn.linear_model import LinearRegression as lr
```

```python
img.shape
```

```python
LR = lr()
LR.fit(X=data_set,y=img.flatten())
```

```python
y_hat= np.matmul(data_set,LR.coef_)
```

```python
#y_hat = np.matmul(data_set,beta)
plt.subplot(1,2,1)
plt.imshow(y_hat.reshape((s,s)).astype(np.float32)+LR.intercept_)
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(img.reshape((s,s)))
plt.axis('off')
```
